```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException
import pandas as pd
import time
# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open the webpage with the table
url = 'http://dgdagov.info/index.php/pharmacies/allopathic-retail-pharmacy'
driver.get(url)

# Initialize an empty list to store table data
table_data = []

while True:
    try:
        # Find the tbody element within the table
        tbody = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, '#gridData tbody'))
        )
        # Find all rows in the tbody
        rows = tbody.find_elements(By.TAG_NAME, 'tr')

        # Extract data from each row in the tbody
        for row in rows:
            # Find all cells in the row
            cells = row.find_elements(By.TAG_NAME, 'td')
        
            # Extract text from each cell and append to row_data list
            row_data = [cell.text for cell in cells]
            print(row_data)
        
            # Append row_data to table_data list
            table_data.append(row_data)
            
        next_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, 'gridData_next'))
        )
        if 'disabled' in next_button.get_attribute('class'):
            break
        else:
            next_button.click()
    except StaleElementReferenceException:
        print('StaleElementReferenceException. Retrying...')
        time.sleep(0.0001)
        continue
    
# Close the WebDriver
# driver.quit()

# Convert the table_data list into a pandas DataFrame
df = pd.DataFrame(table_data)

# Display the DataFrame
df.head(10)
```

```python
print(len(df))

# Define column titles
column_titles = ['SL', 'District', 'Upazila/Thana', 'Pharmacy Name', 'License No', 'Date of Renewal', 'Valid Upto', 'More']

# Convert the table_data list into a pandas DataFrame with column titles
df = pd.DataFrame(table_data, columns=column_titles)

# Drop the last column from the DataFrame
df2 = df.drop('More', axis=1)

# Fix index
df2.reset_index()
df2.set_index('SL', inplace=True)

# Save DataFrame to CSV
df2.to_csv('Allpathic Retail Pharmacy.csv', index=False)

df2.head(2)
```
